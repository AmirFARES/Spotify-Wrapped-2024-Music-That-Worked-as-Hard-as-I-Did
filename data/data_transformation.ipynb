```python
import pandas as pd

# Read the JSON file into a DataFrame
json_file_path = "Streaming_History_Audio_2020-2025.json" 
df = pd.read_json(json_file_path)
```

```python
# Rename columns

column_renames = {
    "spotify_track_uri": "spotify_track_uri",
    "ts": "ts",
    "platform": "platform",
    "ms_played": "ms_played",
    "master_metadata_track_name": "track_name",
    "master_metadata_album_artist_name": "artist_name",
    "master_metadata_album_album_name": "album_name",
    "reason_start": "reason_start",
    "reason_end": "reason_end",
    "shuffle": "shuffle",
    "skipped": "skipped",
}

df = df[list(column_renames.keys())].rename(columns=column_renames)
```

```python
# Save the DataFrame to a CSV file
csv_file_path = "transformed_data.csv"
df.to_csv(csv_file_path, index=False)

print(f"Data successfully transformed and saved to {csv_file_path}")
```
